```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Ardavaa/GDG-feature-eng-data-prep/refs/heads/main/dataset-penugasan/MBP.csv") #Membaca file csv sesuai Dir Path

df.head(10)
```

# Data Cleaning

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.rename(columns={'Phone Name': 'Phone','Rating ?/5':'Rating' ,'ROM/Storage': 'ROM', 'Back/Rare Camera': 'Back Camera'}, inplace=True)
```

# Data Pre-Processing

```python
drop_col = ['Date of Scraping']
df.drop(columns=drop_col, inplace=True)
```

```python
df['Phone'].unique()
```

```python
df['Phone'] = df['Phone'].str.split().str[0]
df['Phone']
```

```python
df.info()
```

```python
df['Number of Ratings']
```

```python
df['Number of Ratings'] = df['Number of Ratings'].str.replace(',', '').astype(int)
```

```python
df['RAM'].value_counts()
```

```python
ram_to_remove = ["1 GB RAM", "512 MB RAM", "8 MB RAM", "16 MB RAM", "1.5 GB RAM"]
df = df[~df['RAM'].isin(ram_to_remove)]
```

```python
df['RAM'].value_counts()
```

```python
df['RAM'] = df['RAM'].str.split().str[0]
df['RAM'] = pd.to_numeric(df['RAM'])
```

```python
df.info()
```

```python
df['ROM'].value_counts()
```

```python
rom_to_remove = ["8 GB ROM"]
df = df[~df['ROM'].isin(rom_to_remove)]
```

```python
df['ROM'] = df['ROM'].str.split().str[0]
df['ROM'] = pd.to_numeric(df['ROM'], errors='coerce')
```

```python
df['ROM'].value_counts()
```

```python
df
```

```python
df['Price in INR'] = df['Price in INR'].str.replace('₹', '').str.replace(',', '').astype(float)
```
